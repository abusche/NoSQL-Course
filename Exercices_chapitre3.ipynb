```python
import tqdm
import json
import pymongo

client = pymongo.MongoClient('localhost', 27017)
mydb = client["ex_chap3"]
collection = mydb["RWA1"]

with open("C:/Users/busch/OneDrive/Documents/Fac/M1/UE16-IDD/NoSQL/pubmed_cleaned.json", 'r') as fj:
    data=json.load(fj)

dt = []   
for i in  tqdm.tqdm(range(len(data))):
    t = data[i]
    cle = list(t["_id"].keys())[0]
    nouvelle_cle = cle.replace("$","")
    t["_id"][nouvelle_cle] = t["_id"].pop(cle)
    dt.append(t)

collection.insert_many(dt)
```
